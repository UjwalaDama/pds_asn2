```python
import pandas as pd
import re
```

```python
rawd=pd.read_csv("/train.csv")
```

```python
#Q1. deleting all the missing values
rawd.dropna()
rawd.dropna(subset=['New_Price'], inplace=True)
rawd.dropna(subset=['Mileage'], inplace=True)
rawd.dropna(subset=['Power'], inplace=True)
```

```python
rawd.isnull().sum()
#Handling the missing values
```

```python
#Q2 Take away some of the attributes' units and retain merely the number values
rawd['Mileage'] = rawd['Mileage'].apply(lambda x: re.sub(r'[^\d.]', '', str(x)))
rawd['Engine'] = rawd['Engine'].apply(lambda x: re.sub(r'[^\d.]', '', str(x)))
rawd['Power'] = rawd['Power'].apply(lambda x: re.sub(r'[^\d.]', '', str(x)))
rawd['New_Price'] = rawd['New_Price'].apply(lambda x: re.sub(r'[^\d.]', '', str(x)))
rawd.head()
```

```python
rawd= pd.get_dummies(rawd, columns=['Fuel_Type', 'Transmission'])
print(rawd.dtypes)
```

```python
#Q4.Creating one more feature and here we are showing the curent car age
from sklearn.preprocessing import LabelEncoder
current_year = pd.to_datetime('today').year
rawd['Car_Age'] = current_year - rawd['Year']
rawd.head()
```
